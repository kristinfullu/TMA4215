## Task 1: Properties of Bernstein polynomials

We consider Bernstein polynomials,

$$ B_{i, n}(t) = {n \choose i} t^i (1 - t)^{n - i}, i = 0, ..., n, $$

where $n \in \mathbb{N}$.

### a)

Want to show that $B_{i, n}(t) \geq 0$ for all $t \in [0,1]$:
Using induction on $n$ to show this with the recursive definition of the Bernstein polynomials, 

$$ B_{i, k}(t) = (1 - t)B_{i, k - 1}(t) + tB_{i - 1, k - 1}(t), k \in \mathbb{N}. $$

Kilde: https://www.final.edu.tr/docs/10118972894pdf[1546613790].pdf?fbclid=IwAR3-XyzcTYTFrFQcYYudXthZVdcEkiUsDiKCMW-lZZWiPE_LBSIq1pdDofg

First, we have that for $n = 0$

$$ B_{0, 0}(t) = {0 \choose 0} t^0 (1 - t)^{0 - 0} = 1 \geq 0 $$

for all $t \in [0, 1]$. As the inductive hypothesis, we assume that all Bernstein polynomials, $B_{i, k - 1}(t) \geq 0$ for all $t \in [0, 1]$. Then, we want to show that the case $k$ also is non-negative using the recursive definition. We have that $(1 - t) \geq 0$ and $t \geq 0$ for all $t \in [0, 1]$. The Bernstein polynomials, $B_{i, k - 1}(t)$ and $B_{i - 1, k - 1}(t)$, are also non-negative from the inductive hypothesis. Then $B_{i, k}(t) \geq 0$ for all $t \in [0, 1]$, since every component in the recursive definition is non-negative. 



### b)

Now we want to show that the Bernstein polynomials can recursively be defined by

$$ B_{i, n}(t) = (1 - t)B_{i, n - 1}(t) + tB_{i - 1, n - 1}(t), t \in [0, 1], $$

for $0 < i < n$. We have that

$$ B_{i, n - 1}(t) = {n - 1 \choose n} t^i (1 - t)^{n - 1 - i} $$ 

and 

$$ B_{i - i, n - 1}(t) = {n - 1 \choose i - 1} t^{i - 1} (1 - t)^{n - 1 - i - 1} = {n - 1 \choose i - 1} t^{i - 1} (1 - t)^{n - i} . $$

Writing out the right-hand side of the recursive definition:

$$ (1 - t){n - 1 \choose n} t^i (1 - t)^{n - 1 - i} + t {n - 1 \choose i - 1} t^{i - 1} (1 - t)^{n - i} = {n - 1 \choose n} t^i (1 - t)^{n - 1 - i + 1} + {n - 1 \choose i - 1} t^{i - 1 + 1} (1 - t)^{n - i} $$ 

$$ = {n - 1 \choose n} t^i (1 - t)^{n - i} + {n - 1 \choose i - 1} t^i (1 - t)^{n - i} = ({n - 1 \choose n} + {n - 1 \choose i - 1}) t^i (1 - t)^{n - i} $$

Using the reccurence relation, $ {n \choose i} = {n - 1 \choose n} + {n - 1 \choose i - 1}$:

Kilde: https://en.wikipedia.org/wiki/Binomial_coefficient 

$$ ({n - 1 \choose n} + {n - 1 \choose i - 1}) t^i (1 - t)^{n - i} = {n \choose i} t^i (1 - t)^{n - i} = B_{i, n}(t) $$

Then the Bernstein polynomials have the recursive definition, 

$$ B_{i, n}(t) = (1 - t)B_{i, n - 1}(t) + tB_{i - 1, n - 1}(t). $$



### c)

Now, we want to show that the $B_{i,n}(t)$ from a partition of unity, i.e

$$
\sum_{i=0}^{n}B_{i,n}(t) = 1
$$

We will prove this by induction of n. We will need the following definitions, where we proved the second one in b).

$$
B_{i,n}(t) = \begin{pmatrix} n \\ i \end{pmatrix} t^{i}(1-t)^{n-i} 
\\
B_{i,n}(t) = (1-t)B_{i,n-1}(t) + tB_{i-1,n-1}(t)
$$

First, we will check if it's true for $n=1$:

$$
\sum_{i=0}^{1}B_{i,1}(t) = B_{0,1}(t) + B_{1,1}(t)
$$

Lets use the first definition to calculate this

$$
\begin{split}
\sum_{i=0}^{1}B_{i,1}(t) &= \begin{pmatrix} 1 \\ 0 \end{pmatrix} t^{0}(1-t)^{1-0}  + \begin{pmatrix} 1 \\ 1 \end{pmatrix} t^{1}(1-t)^{1-1} \\
& = \frac{1!}{0!(1-0)!}(1-t) + \frac{1!}{1!(1-1)!}t \\
& = 1-t + t 
\\
& = 1
\end{split}
$$

Now lets assume that it is true for $n=k$

$$
\sum_{i=0}^{k}B_{i,k}(t) = 1
$$

We want to use this assumption to show that it's also true for $n=k+1$

$$
\sum_{i=0}^{k+1}B_{i,k+1}(t) = B_{0,k+1}(t) + B_{k+1,k+1}(t) + \sum_{i=0}^{k}B_{i,k+1}(t)
$$

Lets use the second definition on the last term

$$
\begin{split}
\sum_{i=0}^{k+1}B_{i,k+1}(t) & = B_{0,k+1}(t) + B_{k+1,k+1}(t) + \sum_{i=1}^{k}((1-t)B_{i,(k+1)-1}(t) + tB_{i-1,(k+1)-1}(t)) \\
& = B_{0,k+1}(t) + B_{k+1,k+1}(t) + \sum_{i=1}^{k}((1-t)B_{i,k}(t) + tB_{i-1,k}(t)) \\
& = B_{0,k+1}(t) + B_{k+1,k+1}(t) + (1-t)\sum_{i=1}^{k}B_{i,k}(t) + t\sum_{i=1}^{k}B_{i-1,k}(t) \\
& = B_{0,k+1}(t) + B_{k+1,k+1}(t) + (1-t)(\sum_{i=0}^{k}B_{i,k}(t) - B_{0,k}(t)) + t\sum_{i=1}^{k}B_{i-1,k}(t) 
\end{split}
$$

Lets first calculate the two first terms and the $B_{0,k}(t)$ term. 

$$
B_{0,k+1}(t) = \frac{(k+1)!}{0!(k+1)!} t^0 (1-t)^{k+1} = (1-t)^{k+1}
\\
\;
\\
B_{k+1,k+1}(t) = \frac{(k+1)!}{(k+1)!(k+1 - (k+1))!} t^{k+1}(1-t)^0 = t^{k+1}
\\
\;
\\
B_{0,k}(t) = \frac{k!}{0!k!}t^0 (1-k)^k = (1-t)^k
$$

Lets also calculate the last term, by shifting the index, and use the assumption

$$
\begin{split}
\sum_{i=1}^{k}B_{i-1,k}(t) & = \sum_{j=0}^{k-1}B_{j,k}(t) + B_{k,k}(t) - B_{k,k}(t) \\
& = \sum_{j=0}^{k}B_{j,k}(t) - B_{k,k}(t) = 1 - t^k
\end{split}
$$

Now, lets put everything together

$$
\begin{split}
\sum_{i=0}^{k+1}B_{i,k+1}(t) & = (1-t)^{k+1} + t^{k+1} + (1-t)(1-(1-t)^k) + t(1-t^k) \\
& = (1-t)^{k+1} + t^{k+1} + (1-t) - (1-t)^{k+1} + t - t^{k+1} \\
& = (1-t)^{k+1} + t^{k+1} + 1-t - (1-t)^{k+1} + t - t^{k+1} \\
& = 1
\end{split}
$$

Thus, we have shown that the $B_{i,n}$ in fact form a unity partition.

### e) 

Implementing a function Bernstein$(i, n, t)$ that evaluates $B_{i, n}$ at $t$:

```python
def Bernstein(i, n, t):
    """
    Evaluating Bernstein polynomial.

    Input: 
    i - integers from 0 to n
    n - degree of the Bernstein polynomial
    t - time 

    Output:
    binom_coeff*t**i*(1- t)**(n - i) - Bernstein polynomial
    """
    binom_coeff = scipy.special.binom(n, i)
    return binom_coeff*t**i*(1 - t)**(n - i)

# Plotting all functions B_{i, n}, i = 0, ..., n for n = 4, 5, 15:

# t-values
t = np.linspace(0, 1, 100)

def plot_Bernstein(n):
    # Sub plots
    fig, axes = plt.subplots(figsize = (8, 6))

    # Labels
    plt.title(f'Bernstein polynomials, n = {n} ', fontsize = 18)
    plt.xlabel("t")
    plt.ylabel(f'B_i,{n}')

    for i in range(0,n + 1):
        y = Bernstein(i, n, t)
        B = axes.plot(t, y)

    plt.show()

plot_Bernstein(4)
plot_Bernstein(5)
plot_Bernstein(15)

```

### d)

We want to use the recursion property for the Bernstein polynomials to prove the recursive definition

$$
\vec{b_n}(t;\vec{p_0}, \ldots, \vec{p_n}) = (1-t)\vec{b}_{n-1}(t;\vec{p_0}, \ldots, \vec{p}_{n-1}) + t\vec{b}_{n-1}(t;\vec{p_1}, \ldots, \vec{p_n})
$$

To prove this, we need the following definition

$$
\vec{b_n}(t;\vec{p_0}, \ldots, \vec{p_n}) = \sum_{i=0}^{n}B_{i,n}(t)\vec{p}_i
$$

We also need the following that we proved in 1b)

$$
B_{i, n}(t) = (1 - t)B_{i, n - 1}(t) + tB_{i - 1, n - 1}(t)
$$

Now lets use the to prove our statement

$$
\begin{split}
\vec{b_n}(t;\vec{p_0}, \ldots, \vec{p_n}) & = \sum_{i=0}^{n}B_{i,n}(t)\vec{p}_i 
\\
& = B_{0,n}(t)\vec{p_0} + \sum_{i=1}^{n-1}B_{i,n}(t)\vec{p}_i + B_{n,n}(t)\vec{p_n} 
\\
& = B_{0,n}(t)\vec{p_0} + \sum_{i=1}^{n-1}((1 - t)B_{i, n - 1}(t) + tB_{i - 1, n - 1}(t))\vec{p_i} + B_{n,n}(t)\vec{p_n} 
\\
& = B_{0,n}(t)\vec{p_0} + \sum_{i=1}^{n-1}(1 - t)B_{i, n - 1}(t)\vec{p_i} + \sum_{i=1}^{n-1} tB_{i - 1, n - 1}(t)\vec{p_i}+ B_{n,n}(t)\vec{p_n} 
\\
& = B_{0,n}(t)\vec{p_0} - (1-t)B_{0,n-1}(t)\vec{p_0} + \sum_{i=0}^{n-1}(1 - t)B_{i, n - 1}(t)\vec{p_i} + \sum_{i=1}^{n} tB_{i - 1, n - 1}(t)\vec{p_i}+ B_{n,n}(t)\vec{p_n} - tB_{n-1,n-1}(t)\vec{p_n}
\end{split}
$$

Lets calculate the two first and the two last terms

$$
B_{0,n}(t)\vec{p_0} = \frac{n!}{0!n!}t^0(1-t)^n\vec{p_0} = (1-t)^n \vec{p_0}
\\
\;
\\
B_{n,n}(t)\vec{p_n} = \frac{n!}{n!0!}t^n(1-t)^0\vec{p_n} = t^n\vec{p_n}
\\
\;
\\
B_{0,n-1}(t)\vec{p_0} = \frac{(n-1)!}{0!(n-1)!}t^0(1-t)^{n-1}\vec{p_0} = (1-t)^{n-1}\vec{p_0}
\\
\;
\\
B_{n-1,n-1}(t)\vec{p_n} = \frac{(n-1)!}{(n-1)!0!}t^{n-1}(1-t)^0\vec{p_n} = t^{n-1}\vec{p_n}
$$

Now, lets put everything in the equation 

$$
\begin{split}
\vec{b_n}(t;\vec{p_0}, \ldots, \vec{p_n}) & = (1-t)^n\vec{p_0} - (1-t)(1-t)^{n-1}\vec{p_0} + (1-t)\sum_{i=0}^{n-1}B_{i, n - 1}(t)\vec{p_i} + t\sum_{i=1}^{n}B_{i-1, n - 1}(t)\vec{p_i} + t^n\vec{p_n} - t \cdot t^{n-1}\vec{p_n}
\\
& = (1-t)^n\vec{p_0} - (1-t)^n\vec{p_0} + (1-t)\sum_{i=0}^{n-1}B_{i, n - 1}(t)\vec{p_i} + t\sum_{i=1}^{n}B_{i-1, n - 1}(t)\vec{p_i} + t^n\vec{p_n} - t^n\vec{p_n} 
\\
& = \sum_{i=0}^{n-1}B_{i, n - 1}(t)\vec{p_i} + t\sum_{i=1}^{n}B_{i-1, n - 1}(t)\vec{p_i} 
\\
& = (1-t)\vec{b}_{n-1}(t;\vec{p_0}, \ldots, \vec{p}_{n-1}) + t\vec{b}_{n-1}(t;\vec{p_1}, \ldots, \vec{p_n})
\end{split}
$$

Thus the statement is proved.

```python
def de_Casteljau(P,t):
    '''
    Input:
        P: a set of control points
        t: time steps
    Output:
        p_vec[0]: Bezier curve at time t, the final level is the point of the curve
        Pvecs: a list of intermediate points at each level
    '''
    p_vec = [p for p in P]
    Pvecs = []
    n = len(p_vec) # degree of the Bezier curve

    for k in range(1,n):
        level_vec = []
        for i in range(n-k):
            # Recursion
            x = (1-t)*p_vec[i][0] + t*p_vec[i+1][0] #linearly interpolating between points for x -coord
            y = (1-t)*p_vec[i][1] + t*p_vec[i+1][1] #linearly interpolating between points for y -coord

            # add to level vector
            level_vec.append((x,y))

            # update control point
            p_vec[i] = (x,y)

        Pvecs.append(level_vec)

    return p_vec[0], Pvecs
```

```python
# Control points
p0 = np.array([0,0])
p1 = np.array([-1,5])
p2 = np.array([4,4])
p3 = np.array([9,3])
p4 = np.array([8,8])

P = [p0, p1, p2, p3, p4]

# Define t0
t0 = np.linspace(1/3,1,1000)

# Call the function with a list of control points
b, Pvecs = de_Casteljau(P, t0)

# Plot the Bézier curve
plt.plot(b[0], b[1], label='Bézier Curve') # gives us the shape of the curve

# # Plot the intermediate points
# for level_points in Pvecs:
#     plt.plot(*zip(*level_points), marker='x', linestyle='--', color='gray') 

# Plot the control points
plt.scatter(*zip(*[p0, p1, p2, p3, p4]), color='red', label='Control Points') 

plt.title("Bezier curve with de Casteljau")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```

#### 1) 

The critical/interesting points of $\mathbf{s}(t)$ to investigate for it to be a $C^k$ function is at $t = 1$ for $C^0$, $C^1$ and $C^2$. That is, we want to check for continuity in $s(1)$, $s'(1)$ and $s''(1)$.

Kilde: http://math.caltech.edu/~syye/teaching/courses/Ma8_2015/Lecture%20Notes/ma8_wk6.pdf

#### 2) 

For continuity ($k = 0$) to hold, $c(1) = d(0)$ has to hold. 

#### 3) 

For $\mathbf{s}(t)$ to be continuous differentiable ($k = 1$), $2)$, and $c'(1) = d'(0)$ have to hold.

#### 4)

For $\mathbf{s}(t)$ to be twice continuous differentiable ($k = 2$), $3)$, and $c''(1) = d''(0)$ have to hold. 

#### 5) 

Let $n = 3$. Degrees of freedom we have left to interpolate the data points $\mathbf{d}_0$, $\mathbf{d}_1$, $\mathbf{d}_2 \in \mathbb{R}^d$ with $\mathbf{s}(t)$, 

$$\mathbf{s}(h) = \mathbf{d}_h, h = 0, 1, 2$$ 

for $\mathbf{s} \in C^1$ is in other words, how many of the points $\mathbf{p}_0, ..., \mathbf{p}_3, \mathbf{q}_0, ..., \mathbf{q}_3$ we can choose freely. For $\mathbf{s}$ to be a $C^1$ function, then $3)$ has to hold. From $2b)$, 

$$\mathbf{c}(1) = \mathbf{p}_3 = \mathbf{q}_0 = \mathbf{d}(0).$$

Then we can choose one of these point freely, but the other has to equal. From $2c)$, 

$$\mathbf{c}'(1) = 3(\mathbf{p}_3 - \mathbf{p}_2) = 3(\mathbf{q}_1 - \mathbf{q}_0) = \mathbf{d}'(0).$$

We already know that $\mathbf{p}_3 = \mathbf{q}_0$, then we can choose one of $\mathbf{p}_2$ and $\mathbf{q}_1$ freely, and then the other one is set. Then there is only $2$ points we can't choose freely, that is, there is $6$ points we can choose freely. Then the degrees of freedom is $6$.

Now, we consider $\mathbf{s} \in C^2$. Then $4)$ has to hold. From $2c)$,

$$\mathbf{c}''(1) = 3(3 - 1)(\mathbf{p}_3 - 2\mathbf{p}_2 + \mathbf{p}_1) = 3(3 - 1)(\mathbf{q}_0 - 2\mathbf{q}_1 + \mathbf{q}_2) = \mathbf{d}''(0).$$

In addition to the $2$ points from the case $k = 1$, we now also have this last condition where we can choose one of the points $\mathbf{p}_1$ and $\mathbf{q}_2$ freely and the other one is then set. Then there is one less point we can choose freely than in case $k = 1$. Thus the degrees of freedom is $5$. 

#### 6)

If we want $\mathbf{s}(t)$ to be periodic, i.e. $\mathbf{s}(t) = \mathbf{s}(t + 2)$ for all $t$ for the continuously differentiability case $k = 1$, then $\mathbf{s}'(t) = \mathbf{s}'(t + 2)$ has to hold. That is, $\mathbf{c}(0) = \mathbf{d}(1)$ and $\mathbf{c}'(0) = \mathbf{d}'(1)$ have to hold. Then for the degrees of freedom for the kase $k = 1$ we now also have these conditions, 

$$\mathbf{c}(0) = \mathbf{p}_0 = \mathbf{q}_3 = \mathbf{d}_1$$

and 

$$\mathbf{c}'(0) = 3(\mathbf{p}_1 - \mathbf{p}_0) = 3(\mathbf{q}_3 - \mathbf{q}_2) = \mathbf{d}'(1).$$

Then we can choose one of the points $\mathbf{p}_0$ and $\mathbf{q}_3$ and one of the points $\mathbf{p}_1$ and $\mathbf{q}_2$ freely. Then there are $2$ less points we can choose freely than in our original case $k = 1$. Thus the degrees of freedom is $4$.


## Task 3: Interpolation with (composite, cubic) Bézier curves

```python
def plotBezier(p,t,title="Composite Bezier curve"):
    '''A function to plot the composite Bezier curves'''
    
    xvals, yvals = composite_Bezier(p,t) # Computes the curve

    for i in range(np.size(p,axis=0)): # Plot each spline by itself
        plt.plot(xvals[i], yvals[i])

    # Extract the interpolation points by themselves
    q = p.reshape(int(np.size(p)/2),2) 
    x = q[:,0]
    y = q[:,1]

    #Plot
    plt.title(title, fontsize = 20)
    plt.xlabel("x", fontsize = 15)
    plt.ylabel("y", fontsize = 15)
    plt.plot(x,y,'o',color="teal")
    plt.grid(True)
    plt.show() 
```

```python
def composite_Bezier(p,t):
    '''
    Input: Matrix P of control points, time t
    Output: Matrix of evaluated B(t)
    '''
    
    i = np.size(p,axis=0) # Number of splines
    j = np.size(p,axis=1) # Number of points per spline
    n = j-1               # Degree of polynomial
    
    # Matrices to store the computed values
    xVals = np.zeros((i,len(t)))
    yVals = np.zeros((i,len(t)))
    
    for s in range(i): # Iterating through the splines
        spline = p[s:s+1,:,:][0]# 0 index added to remove extra parentheses
        x = spline[:,0] # Extract the x and y values respectively
        y = spline[:,1]
        
        xTemp = np.zeros_like(t)
        yTemp = np.zeros_like(t)
        
        for js in range(j): # Computes the sum in the expression
            xTemp = np.add(xTemp, Bernstein(js, n, t)*x[js])
            yTemp = np.add(yTemp, Bernstein(js, n, t)*y[js])
        xVals[s] = xTemp
        yVals[s] = yTemp
    
    return xVals, yVals
```

### b)

Using the properties derived so far to derive an algorithm for the following problem:

Given $\mathbf{a}_0, ..., \mathbf{a}_{m - 1} \in \mathbb{R}^d$ as our data points, and $\mathbf{v}_0, ..., \mathbf{v}_{m - 1} \in \mathbb{R}^d$ as our velocities we want to find the periodic composite cubic Bézier curve $\mathbf{B}(t)$ that maps from $[0, m]$ to $\mathbb{R}^d$ with the following properties:

- $\mathbf{B}(0) = \mathbf{B}(m)$, which means that $\mathbf{a}_0 = \mathbf{a}_{m}$ has to hold. 
- $\mathbf{B}'(0) = \mathbf{B}'(m)$, which means that $\mathbf{v}_0 = \mathbf{v}_{m}$ has to hold.
- $\mathbf{B}(i) = \mathbf{a}_i$ for $i = 0, ..., m - 1$,
- $\mathbf{B}'(i) = \mathbf{v}_i$ for $i = 0, ..., m - 1$

To find this $\mathbf{B}(t)$, we need to find the control points $P = (\mathbf{p}_{j, i})^{3, m}_{j = 0, i = 1}$. To compute the control points in each segment $i$ we use our given data points and velocities. We want to add two control points between each $\mathbf{a}_i$ points. Let $\mathbf{p}_{0, 1} = \mathbf{a}_0$ and $\mathbf{p}_{3, 1} = \mathbf{a}_1$, then using what the properties from task $2c)$ we have that,

$$\mathbf{v}_0 = \mathbf{B}'(0) = 3(\mathbf{p}_{1, 1} - \mathbf{p}_{0, 1}) = 3(\mathbf{p}_{1, 1} - \mathbf{a}_0) $$

and 

$$\mathbf{v}_1 = \mathbf{B}'(1) = 3(\mathbf{p}_{3, 1} - \mathbf{p}_{2, 1}) = 3(\mathbf{a}_1 - \mathbf{p}_{2, 1}).$$

Then 

$$ \mathbf{p}_{1, 1} = \frac{\mathbf{v}_0 + 3\mathbf{a}_0}{3}$$

and 

$$ \mathbf{p}_{2, 1} = \frac{3\mathbf{a}_1 - \mathbf{v}_1}{3}.$$

If we do this for each segment $i$, we have derived all our control points. For segment $i = m$, we want to ensure periodicity. To do this, let $\mathbf{p}_{1, m} = \mathbf{a}_{m - 1}$, $\mathbf{p}_{3, m} = \mathbf{p}_{0, 1}$, and $\mathbf{v}_{m} = \mathbf{v}_{0}$ to derive the last control points. 

Further, we implement a function interpolate_periodic$(A, V)$, where $A$ is the vector of interpolation points $[\mathbf{a}_0, ..., \mathbf{a}_{m - 1}]$ and $V$ is the vector of the velocities $[\mathbf{v}_0, ..., \mathbf{v}_{m - 1}]$.

```python
def interpolate_periodic(A, V):
    """
    Computing the control points to for a periodic composite cubic Bézier curve.

    Input:
    A - Vector of interpolation points
    V - Vector of velocities

    Output:
    P - Matrix of the control points
    """
    Ptot = [] # Empty list to store points in
    P = [] #Empty list for each spline
    for i in range(len(A)-1): # Handles each case except the last one
        P1 = (V[i] + 3*A[i])/3
        P2 = (3*A[i + 1] - V[i + 1])/3
        P.append(A[i])
        P.append(P1)
        P.append(P2)
        P.append(A[i+1])
        Ptot.append(P)
        P = []
    
    #Last case
    P1 = (V[-1] + 3*A[-1])/3
    P2 = (3*A[0] - V[0])/3
    P.append(A[-1])
    P.append(P1)
    P.append(P2)
    P.append(A[0])
    Ptot.append(P)
    P = []
            
    return np.array(Ptot)
```

Since $\mathbf{B}(t)$ is continously differentiable, we have that $\mathbf{p}_{3, i} = \mathbf{p}_{0, i + 1}$ and $\mathbf{p}_{2, i} = 2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}$. Then we can write

$$\tilde{F}(\mathbf{p}_{0, i}, \mathbf{p}_{1, i}, \mathbf{p}_{2, i}, \mathbf{p}_{3, i}) = \tilde{F}(\mathbf{p}_{0, i}, \mathbf{p}_{1, i}, 2\mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}, \mathbf{p}_{0, i + 1}) = \int_0^1||\mathbf{b}_3''(t;\mathbf{p}_{0, i}, \mathbf{p}_{1, i}, 2\mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}, \mathbf{p}_{0, i + 1})||_2^2 \text{d}t $$

which holds for each segment $i = 1, ..., m - 1.$ For segment $m$, we do not have a segment $m + 1$ to express our points with. Then for this segment

$$\tilde{F}(\mathbf{p}_{0, m}, \mathbf{p}_{1, m}, \mathbf{p}_{2, m}, \mathbf{p}_{3, m}) = \int_0^1||\mathbf{b}_3''(t;\mathbf{p}_{0, m}, \mathbf{p}_{1, m}, \mathbf{p}_{2, m}, \mathbf{p}_{3, m})||_2^2 \text{d}t. $$

Then we can express $F$ as

$$F(\mathbf{P}) = \sum^{m - 1}_{i = 1} \tilde{F}(\mathbf{p}_{0, i}, \mathbf{p}_{1, i}, 2\mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}, \mathbf{p}_{0, i + 1}) + \tilde{F}(\mathbf{p}_{0, m}, \mathbf{p}_{1, m}, \mathbf{p}_{2, m}, \mathbf{p}_{3, m}).$$

Thus $F$ is minimized to the remaining control points $\mathbf{P} = [\mathbf{p}_{0, 1}, \mathbf{p}_{1, 1}, \mathbf{p}_{0, 2}, \mathbf{p}_{1, 2}, ..., \mathbf{p}_{0, m - 1}, \mathbf{p}_{1, m - 1}, \mathbf{p}_{0, m}, \mathbf{p}_{1, m}, \mathbf{p}_{2, m}, \mathbf{p}_{3, m}]$ and we can use the gradient of $\tilde{F}$ to compute the gradient of $F$.





$$\nabla F(\mathbf{P}) = (\nabla_{\mathbf{p}_{0, 1}}F, \nabla_{\mathbf{p}_{1, 1}}F, \nabla_{\mathbf{p}_{0, 2}}F, \nabla_{\mathbf{p}_{1, 2}}F, ..., \nabla_{\mathbf{p}_{0, m - 1}}F, \nabla_{\mathbf{p}_{1, m - 1}}F, \nabla_{\mathbf{p}_{0, m}}F, \nabla_{\mathbf{p}_{1, m}}F, \nabla_{\mathbf{p}_{2, m}}F, \nabla_{\mathbf{p}_{3, m}}F) $$

Lets write a general expression for $\nabla \tilde{F}$ using $\mathbf{p}_{3, i} = \mathbf{p}_{0, i + 1}$ and $\mathbf{p}_{2, i} = 2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}$,

$\nabla \tilde{F} = 
\begin{bmatrix}

-36 \mathbf{p}_{1, i} + 108 \mathbf{p}_{0, i + 1} - 72 \mathbf{p}_{1, i + 1}\\
12 \mathbf{p}_{0, i} - 48 \mathbf{p}_{0, i + 1} + 36 \mathbf{p}_{1, i + 1}
\end{bmatrix}^T$

Then 

$\nabla F(\mathbf{P}) = 
\begin{bmatrix}

### c)

Now, we look at the first order optimality conditions to minimize $\tilde{F}$. If $\mathbf{x}^*$ is a local minimizer, then $\nabla \tilde{F}(\mathbf{x}^*) = 0$.

$$24\mathbf{q}_0 - 36\mathbf{q}_1 + 12\mathbf{q}_3 = 0$$

$$-36\mathbf{q}_0 + 72\mathbf{q}_1 - 36\mathbf{q}_2 = 0$$

$$-36\mathbf{q}_1 + 72\mathbf{q}_2 - 36\mathbf{q}_3 = 0$$

$$12\mathbf{q}_0 - 36\mathbf{q}_2 + 24\mathbf{q}_3 = 0$$

Solving this, we have that $\mathbf{q}_0 = 2\mathbf{q}_2 - 2\mathbf{q}_3$ and $\mathbf{q}_1 = 2\mathbf{q}_2 - \mathbf{q}_3$. Then our minimizer is $\mathbf{x}^* = [2\mathbf{q}_2 - 2\mathbf{q}_3, 2\mathbf{q}_2 - \mathbf{q}_3, \mathbf{q}_2, \mathbf{q}_3]$. This means that we can choose $\mathbf{q}_2$ and $\mathbf{q}_3$ freely, but $\mathbf{q}_0$ and $\mathbf{q}_1$ are then given. Then $\nabla \tilde{F}$ has $2$ degrees of freedom and therefore the solution is not unique. 

Further, we consider the overall problem $F$. Now, we also have to take into account the properties between segments from $C^1$, $\mathbf{p}_{3, i} = \mathbf{p}_{0, i + 1}$ and $\mathbf{p}_{2, i} = 2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}$. Then our minimizer for each segment is 

$$\mathbf{x}^*_i = [2\mathbf{p}_{2, i} - 2\mathbf{p}_{3, i}, 2\mathbf{p}_{2, i} - \mathbf{p}_{3, i}, \mathbf{p}_{2, i}, \mathbf{p}_{3, i}] = [2(2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}) - 2\mathbf{p}_{0, i + 1}, 2(2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}) - \mathbf{p}_{0, i + 1}, 2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}, \mathbf{p}_{0, i + 1}]$$

$$ = [2 \mathbf{p}_{0, i + 1} - 2\mathbf{p}_{1, i + 1}, 3 \mathbf{p}_{0, i + 1} - 2\mathbf{p}_{1, i + 1}, 2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}, \mathbf{p}_{0, i + 1}], i = 1, ..., m - 1.$$

Then 

$\mathbf{x}^* = 
\begin{pmatrix}
2 \mathbf{p}_{0, i + 1} - 2\mathbf{p}_{1, i + 1} \\ 
3 \mathbf{p}_{0, i + 1} - 2\mathbf{p}_{1, i + 1} \\

\mathbf{p}_{0, i + 1} \\

\mathbf{p}_{2, m} \\ 
\mathbf{p}_{3, m}
\end{pmatrix}, i = 1, ..., m - 1.$

We can see that each point in segment $i = 1, ..., m - 1$ is given by the points $\mathbf{p}_{0, i + 1}$ and $\mathbf{p}_{1, i + 1}$. We can choose $\mathbf{p}_{2, m}$ and $\mathbf{p}_{3, m}$ freely and the remaining control points are then given. Thus $\nabla F$ also has $2$ degrees of freedom, and the solutionis nor unique either. 



### d)

Uniqueness = if there are more ways to minimize, or if we can only find one way to minimize.
