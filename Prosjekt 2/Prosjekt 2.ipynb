## Task 1: Properties of Bernstein polynomials

We consider Bernstein polynomials,

$$ B_{i, n}(t) = {n \choose i} t^i (1 - t)^{n - i}, i = 0, ..., n, $$

where $n \in \mathbb{N}$.

### a)

We want to show that $B_{i, n}(t) \geq 0$ for all $t \in [0,1]$:
Using induction on $n$ to show this with the recursive definition of the Bernstein polynomials, 

$$ B_{i, k}(t) = (1 - t)B_{i, k - 1}(t) + tB_{i - 1, k - 1}(t), k \in \mathbb{N}. $$ (Ustaoglu, 2014)

First, we have that for $n = 0$

$$ B_{0, 0}(t) = {0 \choose 0} t^0 (1 - t)^{0 - 0} = 1 \geq 0 $$

for all $t \in [0, 1]$. As the inductive hypothesis, we assume that all Bernstein polynomials, $B_{i, k - 1}(t) \geq 0$ for all $t \in [0, 1]$. Then, we want to show that the case $k$ also is non-negative using the recursive definition. We have that $(1 - t) \geq 0$ and $t \geq 0$ for all $t \in [0, 1]$. The Bernstein polynomials, $B_{i, k - 1}(t)$ and $B_{i - 1, k - 1}(t)$, are also non-negative from the inductive hypothesis. Then $B_{i, k}(t) \geq 0$ for all $t \in [0, 1]$, since every component in the recursive definition is non-negative. 



### b)

Now we want to show that the Bernstein polynomials can recursively be defined by

$$ B_{i, n}(t) = (1 - t)B_{i, n - 1}(t) + tB_{i - 1, n - 1}(t), t \in [0, 1], $$

for $0 < i < n$. We have that

$$ B_{i, n - 1}(t) = {n - 1 \choose n} t^i (1 - t)^{n - 1 - i} $$ 

and 

$$ B_{i - i, n - 1}(t) = {n - 1 \choose i - 1} t^{i - 1} (1 - t)^{n - 1 - i - 1} = {n - 1 \choose i - 1} t^{i - 1} (1 - t)^{n - i} . $$

Writing out the right-hand side of the recursive definition:

$$ (1 - t){n - 1 \choose n} t^i (1 - t)^{n - 1 - i} + t {n - 1 \choose i - 1} t^{i - 1} (1 - t)^{n - i} = {n - 1 \choose n} t^i (1 - t)^{n - 1 - i + 1} + {n - 1 \choose i - 1} t^{i - 1 + 1} (1 - t)^{n - i} $$ 

$$ = {n - 1 \choose n} t^i (1 - t)^{n - i} + {n - 1 \choose i - 1} t^i (1 - t)^{n - i} = ({n - 1 \choose n} + {n - 1 \choose i - 1}) t^i (1 - t)^{n - i} $$

Using the reccurence relation, $ {n \choose i} = {n - 1 \choose n} + {n - 1 \choose i - 1}$: (Wikipedia (1), 2023)

$$ ({n - 1 \choose n} + {n - 1 \choose i - 1}) t^i (1 - t)^{n - i} = {n \choose i} t^i (1 - t)^{n - i} = B_{i, n}(t) $$

Then the Bernstein polynomials have the recursive definition, 

$$ B_{i, n}(t) = (1 - t)B_{i, n - 1}(t) + tB_{i - 1, n - 1}(t). $$



For $B_{0, n}(t)$ and $B_{n, n}(t)$, we can't use this definition. 

$$B_{0, n}(t) = {n \choose 0}t^0(1 - t)^{n - 0} = \frac{n!}{0!(n - 0)!}(1 - t)^n = (1 - t)(1 - t)^{n - 1} = (1 - t)B_{0, n - 1}(t)$$

$$B_{n, n}(t) = {n \choose n}t^n(1 - t)^{n - n} = \frac{n!}{n!(n - n)!}t^n = tt^{n - 1} = t B_{i - 1, n - 1}(t)$$

This means that $B_{-1, n - 1}(t) = 0$ and $B_{n, n - 1}(t) = 0$, which makes sense since Bernstein polynomials are not defined when $i < 0$ and $i > n$ since we can't take the faculty of a negative value.

### c)

Now, we want to show that the $B_{i,n}(t)$ from a partition of unity, i.e

$$
\sum_{i=0}^{n}B_{i,n}(t) = 1
$$

We will prove this by induction of n. We will need the following definitions, where we proved the second one in b).

$$
B_{i,n}(t) = \begin{pmatrix} n \\ i \end{pmatrix} t^{i}(1-t)^{n-i} 
\\
B_{i,n}(t) = (1-t)B_{i,n-1}(t) + tB_{i-1,n-1}(t)
$$

First, we will check if it's true for $n=1$:

$$
\sum_{i=0}^{1}B_{i,1}(t) = B_{0,1}(t) + B_{1,1}(t)
$$

Lets use the first definition to calculate this

$$
\begin{split}
\sum_{i=0}^{1}B_{i,1}(t) &= \begin{pmatrix} 1 \\ 0 \end{pmatrix} t^{0}(1-t)^{1-0}  + \begin{pmatrix} 1 \\ 1 \end{pmatrix} t^{1}(1-t)^{1-1} \\
& = \frac{1!}{0!(1-0)!}(1-t) + \frac{1!}{1!(1-1)!}t \\
& = 1-t + t 
\\
& = 1
\end{split}
$$

Now lets assume that it is true for $n=k$

$$
\sum_{i=0}^{k}B_{i,k}(t) = 1
$$

We want to use this assumption to show that it's also true for $n=k+1$

$$
\sum_{i=0}^{k+1}B_{i,k+1}(t) = B_{0,k+1}(t) + B_{k+1,k+1}(t) + \sum_{i=0}^{k}B_{i,k+1}(t)
$$

Lets use the second definition on the last term

$$
\begin{split}
\sum_{i=0}^{k+1}B_{i,k+1}(t) & = B_{0,k+1}(t) + B_{k+1,k+1}(t) + \sum_{i=1}^{k}((1-t)B_{i,(k+1)-1}(t) + tB_{i-1,(k+1)-1}(t)) \\
& = B_{0,k+1}(t) + B_{k+1,k+1}(t) + \sum_{i=1}^{k}((1-t)B_{i,k}(t) + tB_{i-1,k}(t)) \\
& = B_{0,k+1}(t) + B_{k+1,k+1}(t) + (1-t)\sum_{i=1}^{k}B_{i,k}(t) + t\sum_{i=1}^{k}B_{i-1,k}(t) \\
& = B_{0,k+1}(t) + B_{k+1,k+1}(t) + (1-t)(\sum_{i=0}^{k}B_{i,k}(t) - B_{0,k}(t)) + t\sum_{i=1}^{k}B_{i-1,k}(t) 
\end{split}
$$

Lets first calculate the two first terms and the $B_{0,k}(t)$ term. 

$$
B_{0,k+1}(t) = \frac{(k+1)!}{0!(k+1)!} t^0 (1-t)^{k+1} = (1-t)^{k+1}
\\
\;
\\
B_{k+1,k+1}(t) = \frac{(k+1)!}{(k+1)!(k+1 - (k+1))!} t^{k+1}(1-t)^0 = t^{k+1}
\\
\;
\\
B_{0,k}(t) = \frac{k!}{0!k!}t^0 (1-k)^k = (1-t)^k
$$

Lets also calculate the last term, by shifting the index, and use the assumption

$$
\begin{split}
\sum_{i=1}^{k}B_{i-1,k}(t) & = \sum_{j=0}^{k-1}B_{j,k}(t) + B_{k,k}(t) - B_{k,k}(t) \\
& = \sum_{j=0}^{k}B_{j,k}(t) - B_{k,k}(t) = 1 - t^k
\end{split}
$$

Now, lets put everything together

$$
\begin{split}
\sum_{i=0}^{k+1}B_{i,k+1}(t) & = (1-t)^{k+1} + t^{k+1} + (1-t)(1-(1-t)^k) + t(1-t^k) \\
& = (1-t)^{k+1} + t^{k+1} + (1-t) - (1-t)^{k+1} + t - t^{k+1} \\
& = (1-t)^{k+1} + t^{k+1} + 1-t - (1-t)^{k+1} + t - t^{k+1} \\
& = 1
\end{split}
$$

Thus, we have shown that the $B_{i,n}$ in fact form a unity partition. 

### d)

Now we want to show that the derivative of the Bernstein polynomials is given by

$$
B_{i,n}^{'}(t) = n(B_{i-1,n-1}(t) - B_{i, n-1}(t))
$$

The Bernstein polynomials defined by

$$
B_{i,n}(t) = \frac{n!}{i!(n-i)!}t^i(1-t)^{n-i}
$$

Now, lets derivate this expression using the chain rule and the product rule

$$
\begin{split}
B_{i,n}^{'}(t) & = \frac{d}{dt}(\frac{n!}{i!(n-i)!}t^i(1-t)^{n-i}) \\
& = \frac{n!}{i!(n-i)!}((1-t)^{n-i}\frac{d}{dt}t^i + t^i\frac{d}{dt}(1-t)^{n-i}) \\
& = \frac{n!}{i!(n-i)!}(it^{i-1}(1-t)^{n-i} + t^i(n-i)(1-t)^{n-i-1}) \\
& = \frac{n!}{(i-1)!(n-i)!}t^{i-1}(1-t)^{n-i} + \frac{n!}{i!(n-i-1)!}(1-t)^{n-i-1} \\
& = \frac{n(n-1)!}{(i-1)!(n-i)!}t^{i-1}(1-t)^{n-i} + \frac{n(n-1)!}{i!(n-i-1)!}(1-t)^{n-i-1} \\
& = nB_{i-1,n-1}(t) + nB_{i, n-1}(t) \\
& = n(B_{i-1,n-1}(t) + B_{i, n-1}(t))
\end{split}
$$

### e) 

For further use, we implement a function Bernstein$(i, n, t)$ that evaluates $B_{i, n}$ at $t$:

```python
def Bernstein(i, n, t):
    """
    Evaluating Bernstein polynomial.

    Input: 
    i - integers from 0 to n
    n - degree of the Bernstein polynomial
    t - time 

    Output:
    binom_coeff*t**i*(1- t)**(n - i) - Bernstein polynomial
    """

    binom_coeff = scipy.special.binom(n, i)
    return binom_coeff*t**i*(1 - t)**(n - i)


def plot_Bernstein(n):
    '''
    Function for easy plotting of a Bernstein polynomial
    '''
    # Sub plots
    fig, axes = plt.subplots(figsize = (8, 6))

    # Labels
    plt.title(f'Bernstein polynomials, n = {n} ', fontsize = 18)
    plt.xlabel("t")
    plt.ylabel(f'B_i,{n}')

    for i in range(0,n + 1):
        y = Bernstein(i, n, t)
        B = axes.plot(t, y)

    plt.show()

# Plotting all functions B_{i, n}, i = 0, ..., n for n = 4, 5, 15:

# t-values
t = np.linspace(0, 1, 100)

plot_Bernstein(4)
plot_Bernstein(5)
plot_Bernstein(15)

```

We see that Bernstein polynomials are smooth. As $n$ increases we get more polynomials with more complex functions.

### a)

We want to compute the first two derivatives $\mathbf{c}'(t)$ and $\mathbf{c}''(t)$ of $\mathbf{c}(t)$.

Since we have that

$$
\mathbf{c}(t)= \sum^n_{i=0}B_{i,n}(t)\mathbf{p}_i,

we see that the only component that depends on $t$ is $B_{i,n}(t)$. This therefore leaves us with the expression


$$ \mathbf{c}'(t)= \sum^n_{i=0}B_{i,n}'(t)\mathbf{p}_i$$

We therefore only have to consider $B_{i,n}'(t)$ when performing the differentiation.

From task 1d), we know that

$$
B_{i,n}'(t) = n(B_{i-1,n-1}(t) - B_{i, n-1}(t)).

To obtain the second derivative, we use our expression for the first derivative, and apply what we know from task 1d) to compute the derivative.

$$ 
\mathbf{c}''(t)= \sum^n_{i=0}B_{i,n}''(t)\mathbf{p}_i = \sum^n_{i=0}n(B_{i-1,n-1}'(t) - B_{i, n-1}'(t))\mathbf{p}_i

### b)

Evaluating $\mathbf{c}(t)$ at its endpoints, $t=0,1$ could tell us important information about the curve. We therefore compute the values it attends at the end points.

$$
\mathbf{c}(0)= \sum^n_{i=0}B_{i,n}(0)\mathbf{p}_i = \sum^n_{i=0}{n \choose i} 0^i(1-0)^{n-i}\mathbf{p}_i= \sum^n_{i=0}{n \choose i} 0^i\mathbf{p}_i
$$

and 

$$
\mathbf{c}(1)= \sum^n_{i=0}B_{i,n}(1)\mathbf{p}_i = \sum^n_{i=0}{n \choose i} 1^i(1-1)^{n-i}\mathbf{p}_i = \sum^n_{i=0}{n \choose i} 0^{n-i}\mathbf{p}_i.
$$

Looking at the sums in the above expressions, we notice that the only element of the sum that doesn't equal zero appears when we get a term $0^0$. This means we have to assume $0^0$ isn't undefined, but equals $1$. This leaves us with.

$$
\mathbf{c}(0)= {n \choose 0} 0^0\mathbf{p}_0 = \mathbf{p}_0
$$

and

$$
\mathbf{c}(1)= {n \choose n} 0^0\mathbf{p}_n = \mathbf{p}_n.
$$

Since this expression relates to the interpolation of a curve at time $t$, we can interpret these results as saying that at exact times $t=0$ and $t=1$, the interpolated curve passes through the chosen start and end points $\mathbf{p}_0$ and $\mathbf{p}_n$. This is a condition that is needed for the interpolation to be sufficiently good enough.

### d)

We want to use the recursion property for the Bernstein polynomials to prove the recursive definition

$$
\mathbf{b_n}(t;\mathbf{p_0}, \ldots, \mathbf{p_n}) = (1-t)\mathbf{b}_{n-1}(t;\mathbf{p_0}, \ldots, \mathbf{p}_{n-1}) + t\mathbf{b}_{n-1}(t;\mathbf{p_1}, \ldots, \mathbf{p_n})
$$

To prove this, we need the following definition

$$
\mathbf{b_n}(t;\mathbf{p_0}, \ldots, \mathbf{p_n}) = \sum_{i=0}^{n}B_{i,n}(t)\mathbf{p}_i
$$

We also need the following that we proved in 1b)

$$
B_{i, n}(t) = (1 - t)B_{i, n - 1}(t) + tB_{i - 1, n - 1}(t)
$$

Now lets use the to prove our statement

$$
\begin{split}
\mathbf{b_n}(t;\mathbf{p_0}, \ldots, \mathbf{p_n}) & = \sum_{i=0}^{n}B_{i,n}(t)\mathbf{p}_i 
\\
& = B_{0,n}(t)\mathbf{p_0} + \sum_{i=1}^{n-1}B_{i,n}(t)\mathbf{p}_i + B_{n,n}(t)\mathbf{p_n} 
\\
& = B_{0,n}(t)\mathbf{p_0} + \sum_{i=1}^{n-1}((1 - t)B_{i, n - 1}(t) + tB_{i - 1, n - 1}(t))\mathbf{p_i} + B_{n,n}(t)\mathbf{p_n} 
\\
& = B_{0,n}(t)\mathbf{p_0} + \sum_{i=1}^{n-1}(1 - t)B_{i, n - 1}(t)\mathbf{p_i} + \sum_{i=1}^{n-1} tB_{i - 1, n - 1}(t)\mathbf{p_i}+ B_{n,n}(t)\mathbf{p_n} 
\\
& = B_{0,n}(t)\mathbf{p_0} - (1-t)B_{0,n-1}(t)\mathbf{p_0} + \sum_{i=0}^{n-1}(1 - t)B_{i, n - 1}(t)\mathbf{p_i} + \sum_{i=1}^{n} tB_{i - 1, n - 1}(t)\mathbf{p_i}+ B_{n,n}(t)\mathbf{p_n} - tB_{n-1,n-1}(t)\mathbf{p_n}
\end{split}
$$

Lets calculate the two first and the two last terms

$$
B_{0,n}(t)\mathbf{p_0} = \frac{n!}{0!n!}t^0(1-t)^n\mathbf{p_0} = (1-t)^n \mathbf{p_0}
\\
\;
\\
B_{n,n}(t)\mathbf{p_n} = \frac{n!}{n!0!}t^n(1-t)^0\mathbf{p_n} = t^n\mathbf{p_n}
\\
\;
\\
B_{0,n-1}(t)\mathbf{p_0} = \frac{(n-1)!}{0!(n-1)!}t^0(1-t)^{n-1}\mathbf{p_0} = (1-t)^{n-1}\mathbf{p_0}
\\
\;
\\
B_{n-1,n-1}(t)\mathbf{p_n} = \frac{(n-1)!}{(n-1)!0!}t^{n-1}(1-t)^0\mathbf{p_n} = t^{n-1}\mathbf{p_n}
$$

Now, lets put everything in the equation 

$$
\begin{split}
\mathbf{b_n}(t;\mathbf{p_0}, \ldots, \mathbf{p_n}) & = (1-t)^n\mathbf{p_0} - (1-t)(1-t)^{n-1}\mathbf{p_0} + (1-t)\sum_{i=0}^{n-1}B_{i, n - 1}(t)\mathbf{p_i} + t\sum_{i=1}^{n}B_{i-1, n - 1}(t)\mathbf{p_i} + t^n\mathbf{p_n} - t \cdot t^{n-1}\mathbf{p_n}
\\
& = (1-t)^n\mathbf{p_0} - (1-t)^n\mathbf{p_0} + (1-t)\sum_{i=0}^{n-1}B_{i, n - 1}(t)\mathbf{p_i} + t\sum_{i=1}^{n}B_{i-1, n - 1}(t)\mathbf{p_i} + t^n\mathbf{p_n} - t^n\mathbf{p_n} 
\\
& = \sum_{i=0}^{n-1}B_{i, n - 1}(t)\mathbf{p_i} + t\sum_{i=1}^{n}B_{i-1, n - 1}(t)\mathbf{p_i} 
\\
& = (1-t)\mathbf{b}_{n-1}(t;\mathbf{p_0}, \ldots, \mathbf{p}_{n-1}) + t\mathbf{b}_{n-1}(t;\mathbf{p_1}, \ldots, \mathbf{p_n})
\end{split}
$$

Thus the statement is proved. 

### e)

Implementing a function de_Casteljau$(P, t)$ where $P$ is a vector, of $n + 1$ points, to evaluate the corresponding Bezir curve at $t$, using the recursion formula for the Bernstein polynomials:

```python
def de_Casteljau(P,t):
    '''
    Action: calculates all intermediate points at each level
    Input:
    -P: a set of control points
    -t: time step
    Output:
    -pvecs[-1][0]: Bezier curve at time t
    -Pvecs: a list of intermediate points at each level
    '''
    n = len(P)  # degree of the Bezier curve
    Pvecs = [P] # list of all intermediate points

    # loop for each level
    for k in range(1,n):

        # list to store all intermediate values for each iteration of k
        p_vec = []
        for i in range(n-k):
            # recursive formula for Bernstein
            p_vec.append((1-t)*Pvecs[k-1][i] + t*Pvecs[k-1][i+1])

        Pvecs.append(np.array(p_vec)) # the last values per iteration is added

    return np.array(Pvecs[-1][0]), np.array(Pvecs, dtype=object) # dtype to avoid warning
```

```python
# Define control points
P = np.array([[0,0], [-1,5], [4,4], [9,3], [8,8]])

# Define timesteps
t = np.linspace(0,1,1000)

# Create bezier curve for each timestep
b = []
for t0 in t:
    b.append(de_Casteljau(P, t0)[0])
b = np.array(b)

# Plot the Bzier curve
plt.plot(b[:,0], b[:,1], color="hotpink", label='Bzier Curve') # gives us the shape of the curve

# Evaluate bezier at start-time t=1/3
b_t, Pvecs = de_Casteljau(P, 1/3)

# Plot all levels for this timestep
l = 1
for i in Pvecs:
    plt.plot(i[:,0], i[:,1], linestyle='--')            # lines between intermediate points
    plt.plot(i[:,0], i[:,1], "o", label = f'Level {l}') # intermediate points
    l += 1

plt.title("Bezier curve with de Casteljau")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(fontsize="8")

plt.show()
```

We observe that each level shows how the graph is formed, evaluated at a specific timestep. We observe that there are polygons formed by connecting the Bzier points with lines. These are calles Bzier polygons. The convex hull of the Bzier polygon contains the Bzier curve (Wikipedia (2), 2023).

### a)
In the rest of the project we will be wanting to evaluate points to create a composite Bezier curve. We therefore implement a function composite_Bezier$(p, t)$ that evaluates $\mathbf{B}(t), t \in [0, m]$, where $P = (\mathbf{p}_{j, i})^{3, m}_{j = 0, i = 1}$ denotes a matrix of control points:

```python
def composite_Bezier(p,t):
    '''
    Action: Computes the Bezier curve
    Input:  Matrix P of control points, time t
    Output: Evaluated x- and y-values
    '''
    
    i = np.size(p,axis=0) # Number of splines
    j = np.size(p,axis=1) # Number of points per spline
    n = j-1               # Degree of polynomial
    
    # Matrices to store the computed values
    xVals = np.zeros((i,len(t)))
    yVals = np.zeros((i,len(t)))
    
    for s in range(i): # Iterating through the splines
        spline = p[s:s+1,:,:][0] # 0 index added to remove extra parentheses
        x = spline[:,0] # Extract the x and y values respectively
        y = spline[:,1]
        
        xTemp = np.zeros_like(t)
        yTemp = np.zeros_like(t)
        
        for js in range(j): # Computes the sum in the expression
            xTemp = np.add(xTemp, Bernstein(js, n, t)*x[js])
            yTemp = np.add(yTemp, Bernstein(js, n, t)*y[js])
        xVals[s] = xTemp
        yVals[s] = yTemp
    
    return xVals, yVals
```

### b)

Using the properties derived so far to derive an algorithm for the following problem:

Given $\mathbf{a}_0, ..., \mathbf{a}_{m - 1} \in \mathbb{R}^d$ as our data points, and $\mathbf{v}_0, ..., \mathbf{v}_{m - 1} \in \mathbb{R}^d$ as our velocities we want to find the periodic composite cubic Bzier curve $\mathbf{B}(t)$ that maps from $[0, m]$ to $\mathbb{R}^d$ with the following properties:

- $\mathbf{B}(0) = \mathbf{B}(m)$, which means that $\mathbf{a}_0 = \mathbf{a}_{m}$ has to hold. 
- $\mathbf{B}'(0) = \mathbf{B}'(m)$, which means that $\mathbf{v}_0 = \mathbf{v}_{m}$ has to hold.
- $\mathbf{B}(i) = \mathbf{a}_i$ for $i = 0, ..., m - 1$,
- $\mathbf{B}'(i) = \mathbf{v}_i$ for $i = 0, ..., m - 1$

To find this $\mathbf{B}(t)$, we need to find the control points $P = (\mathbf{p}_{j, i})^{3, m}_{j = 0, i = 1}$. To compute the control points in each segment $i$ we use our given data points and velocities. We want to add two control points between each $\mathbf{a}_i$ points. Let $\mathbf{p}_{0, 1} = \mathbf{a}_0$ and $\mathbf{p}_{3, 1} = \mathbf{a}_1$, then using what the properties from task $2c)$ we have that,

$$\mathbf{v}_0 = \mathbf{B}'(0) = 3(\mathbf{p}_{1, 1} - \mathbf{p}_{0, 1}) = 3(\mathbf{p}_{1, 1} - \mathbf{a}_0) $$

and 

$$\mathbf{v}_1 = \mathbf{B}'(1) = 3(\mathbf{p}_{3, 1} - \mathbf{p}_{2, 1}) = 3(\mathbf{a}_1 - \mathbf{p}_{2, 1}).$$

Then 

$$ \mathbf{p}_{1, 1} = \frac{\mathbf{v}_0 + 3\mathbf{a}_0}{3}$$

and 

$$ \mathbf{p}_{2, 1} = \frac{3\mathbf{a}_1 - \mathbf{v}_1}{3}.$$

If we do this for each segment $i$, we have derived all our control points. For segment $i = m$, we want to ensure periodicity. To do this, let $\mathbf{p}_{0, m} = \mathbf{a}_{m - 1}$, $\mathbf{p}_{3, m} = \mathbf{p}_{0, 1}$, and $\mathbf{v}_{m} = \mathbf{v}_{0}$ to derive the last control points. 

Further, we implement a function interpolate_periodic$(A, V)$, where $A$ is the vector of interpolation points $[\mathbf{a}_0, ..., \mathbf{a}_{m - 1}]$ and $V$ is the vector of the velocities $[\mathbf{v}_0, ..., \mathbf{v}_{m - 1}]$.

```python
def interpolate_periodic(A, V):
    """
    Computing the control points for a periodic composite cubic Bzier curve.

    Input:
    A - Vector of interpolation points
    V - Vector of velocities

    Output:
    P - Vector of all control points, now periodic
    """
    Ptot = [] # Empty list to store points in
    P = []    #Empty list for each spline

    for i in range(len(A)-1): # Handles each case except the last one
        P1 = (V[i] + 3*A[i])/3         
        P2 = (3*A[i + 1] - V[i + 1])/3

        P.append(A[i])   # First point, i.e p_0,i
        P.append(P1)     # p_1,i
        P.append(P2)     # p_2,i
        P.append(A[i+1]) # p_3,i
        Ptot.append(P)
        P = []
    
    #Last case
    P1 = (V[-1] + 3*A[-1])/3
    P2 = (3*A[0] - V[0])/3

    P.append(A[-1])
    P.append(P1)
    P.append(P2)
    P.append(A[0])
    Ptot.append(P)
    P = []
            
    return np.array(Ptot)
```

Bzier curves are used to model smooth curves by a set of control points (Wikipedia (2), 2023). Composite cubic Bezier curves are splines made out of Bzier curves, where each spline is joined end to end and where the last point of one curve coincides with the starting point of the next curve. To add more smootheness to such curves, we need $C^1$ continuity (Wikipedia (3), 2023). Since our curve satisfies $C^1$ continuity, it should be rather smooth. This makes sense, since not only need the curves to be joined end-to-end, but also their first derivatives are continuous at the join point. To sum up, it is continuous in all points, also the first derivatives, such that the curve becomes smooth at the joint point where the curve changes. This ensures that there are no abrupt changes in direction or speed at the connection between two Bzier curves. By looking at it, we observe that the edges are quite smooth, which reinforces this statement.

+12y_0^2-36y_0y_1+36y_1^2-36y_1y_2+36y_2^2+12y_0y_3-36y_2y_3+12y_3^2
$$

Deriving the gradient of 

$$\tilde{F}(\mathbf{q}_0, \mathbf{q}_1, \mathbf{q}_2, \mathbf{q}_3) := \int_0^1||\mathbf{b}_3''(t;\mathbf{q}_0, \mathbf{q}_1, \mathbf{q}_2, \mathbf{q}_3)||_2^2 \text{d}t$$

with respect to the control points $\mathbf{q}_i, i = 0, ..., 3.$

$$\nabla \tilde{F}(\mathbf{q}_0, \mathbf{q}_1, \mathbf{q}_2, \mathbf{q}_3) = (\nabla_{\mathbf{q}_0}\tilde{F}, ..., \nabla_{\mathbf{q}_3}\tilde{F}) = (\frac{\delta \tilde{F}}{\delta x_0}, \frac{\delta \tilde{F}}{\delta y_0} ..., \frac{\delta \tilde{F}}{\delta x_3}, \frac{\delta \tilde{F}}{\delta y_3}) $$

$$= (24x_0 - 36x_1 + 12x_3, 24y_0 - 36y_1 + 12y_3,$$
$$ -36x_0 + 72x_1 - 36x_2, -36y_0 + 72y_1 - 36y_2,$$
$$ -36x_1 + 72x_2 - 36x_3, -36y_1 + 72y_2 - 36y_3,$$
$$ 12x_0 - 36x_2 + 24x_3, 12y_0 - 36y_2 + 24y_3)$$

This can we written together as

$$\nabla \tilde{F} = (24\mathbf{q}_0 - 36\mathbf{q}_1 + 12\mathbf{q}_3,-36\mathbf{q}_0 + 72\mathbf{q}_1 - 36\mathbf{q}_2, -36\mathbf{q}_1 + 72\mathbf{q}_2 - 36\mathbf{q}_3, 12\mathbf{q}_0 - 36\mathbf{q}_2 + 24\mathbf{q}_3)$$

Further, we want to find the gradient of $F$, using the result of $\tilde{F}$. We have that $\mathbf{B}(t)$ is continuosly differentiable, that is, $\mathbf{p}_{3, i} = \mathbf{p}_{0, i + 1}$ and $\mathbf{p}_{2, i} = 2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}$.

Then we can express $F$ as

$$F(\mathbf{P}) = \sum^{m - 1}_{i = 1} \tilde{F}(\mathbf{p}_{0, i}, \mathbf{p}_{1, i}, 2\mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}, \mathbf{p}_{0, i + 1}) + \tilde{F}(\mathbf{p}_{0, m}, \mathbf{p}_{1, m}, \mathbf{p}_{2, m}, \mathbf{p}_{3, m})$$ 

with $\mathbf{P} = [\mathbf{p}_{0, 1}, \mathbf{p}_{1, 1}, \mathbf{p}_{0, 2}, \mathbf{p}_{1, 2}, ..., \mathbf{p}_{0, m - 1}, \mathbf{p}_{1, m - 1}, \mathbf{p}_{0, m}, \mathbf{p}_{1, m}, \mathbf{p}_{2, m}, \mathbf{p}_{3, m}]$.

Writing a general form of $\tilde{F}$ using the $C^1$ conditions:

$$\tilde F = 12\mathbf{p}_{0, i}^2-36\mathbf{p}_{0, i}\mathbf{p}_{1, i}+36\mathbf{p}_{1, i}^2-36\mathbf{p}_{1, i}\mathbf{p}_{2, i}+36\mathbf{p}_{2, i}^2+12\mathbf{p}_{0, i}\mathbf{p}_{3, i}-36\mathbf{p}_{2, i}\mathbf{p}_{3, i}+12\mathbf{p}_{3, i}^2 $$
$$ = 12\mathbf{p}_{0, i}^2-36\mathbf{p}_{0, i}\mathbf{p}_{1, i}+36\mathbf{p}_{1, i}^2-36\mathbf{p}_{1, i}(2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1})+36(2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1})^2+12\mathbf{p}_{0, i}\mathbf{p}_{0, i + 1}-36(2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1})\mathbf{p}_{0, i + 1}+12\mathbf{p}_{0, i + 1}^2$$

If we take the gradient of this, we have that

$$\nabla_{\mathbf{p}_{0, i}} \tilde{F} = 24 \mathbf{p}_{0, i} + 12 \mathbf{p}_{0, i + 1} - 36 \mathbf{p}_{1, i}$$

$$\nabla_{\mathbf{p}_{1, i}} \tilde{F} = -36 \mathbf{p}_{0, i} - 72 \mathbf{p}_{0, i + 1} + 72 \mathbf{p}_{1, i} + 36 \mathbf{p}_{1, i + 1}$$

$$\nabla_{\mathbf{p}_{0, i + 1}} \tilde{F} = 12 \mathbf{p}_{0, i} + 168 \mathbf{p}_{0, i + 1} - 72 \mathbf{p}_{1, i} - 108 \mathbf{p}_{1, i + 1}$$

$$\nabla_{\mathbf{p}_{1, i + 1}} \tilde{F} = -108 \mathbf{p}_{0, i + 1} + 36 \mathbf{p}_{1, i} + 72 \mathbf{p}_{1, i + 1}$$

We see that each segment $i$ rely on points in the next segment $i + 1$. This means that when computing $\nabla F$, we need to take into account that we sum over segments. Lets consider the sum from $i = 1$ to $i = 2$. For $i = 1$:

$$\nabla_{\mathbf{p}_{0, 1}} \tilde{F} = 24 \mathbf{p}_{0, 1} + 12 \mathbf{p}_{0, 2} - 36 \mathbf{p}_{1, 1}$$

$$\nabla_{\mathbf{p}_{1, 1}} \tilde{F} = -36 \mathbf{p}_{0, 1} - 72 \mathbf{p}_{0, 2} + 72 \mathbf{p}_{1, 1} + 36 \mathbf{p}_{1, 2}$$

$$\nabla_{\mathbf{p}_{0, 2}} \tilde{F} = 12 \mathbf{p}_{0, 1} + 168 \mathbf{p}_{0, 2} - 72 \mathbf{p}_{1, 1} - 108 \mathbf{p}_{1, 2}$$

$$\nabla_{\mathbf{p}_{1, 2}} \tilde{F} = -108 \mathbf{p}_{0, 2} + 36 \mathbf{p}_{1, 1} + 72 \mathbf{p}_{1, 2}$$

And for $i = 2$:

$$\nabla_{\mathbf{p}_{0, 2}} \tilde{F} = 24 \mathbf{p}_{0, 2} + 12 \mathbf{p}_{0, 3} - 36 \mathbf{p}_{1, 2}$$

$$\nabla_{\mathbf{p}_{1, 2}} \tilde{F} = -36 \mathbf{p}_{0, 2} - 72 \mathbf{p}_{0, 3} + 72 \mathbf{p}_{1, 2} + 36 \mathbf{p}_{1, 3}$$

$$\nabla_{\mathbf{p}_{0, 3}} \tilde{F} = 12 \mathbf{p}_{0, 2} + 168 \mathbf{p}_{0, 3} - 72 \mathbf{p}_{1, 2} - 108 \mathbf{p}_{1, 3}$$

$$\nabla_{\mathbf{p}_{1, 3}} \tilde{F} = -108 \mathbf{p}_{0, 3} + 36 \mathbf{p}_{1, 2} + 72 \mathbf{p}_{1, 3}$$

We can see that $\nabla_{p_{0, 2}} \tilde{F}$ and $\nabla_{p_{1, 2}} \tilde{F}$ are in both segments. Then 

$$\nabla_{\mathbf{p}_{0, i + 1}} \tilde{F} = 12 \mathbf{p}_{0, i} + 168 \mathbf{p}_{0, i + 1} - 72 \mathbf{p}_{1, i} - 108 \mathbf{p}_{1, i + 1} + 24 \mathbf{p}_{0, i + 1} + 12 \mathbf{p}_{0, i + 2} - 36 \mathbf{p}_{1, i + 1} $$
$$= 12 \mathbf{p}_{0, i} + 192 \mathbf{p}_{0, i + 1} - 72 \mathbf{p}_{1, i} - 144 \mathbf{p}_{1, i + 1} + 12 \mathbf{p}_{0, i + 2}$$

$$\nabla_{\mathbf{p}_{1, i + 1}} \tilde{F} = -108 \mathbf{p}_{0, i + 1} + 36 \mathbf{p}_{1, i} + 72 \mathbf{p}_{1, i + 1} -36 \mathbf{p}_{0, i + 1} - 72 \mathbf{p}_{0, i + 2} + 72 \mathbf{p}_{1, i + 1} + 36 \mathbf{p}_{1, i + 2} $$
$$= -144 \mathbf{p}_{0, i + 1} + 36 \mathbf{p}_{1, i} + 144 \mathbf{p}_{1, i + 1} - 72 \mathbf{p}_{0, i + 2} + 36 \mathbf{p}_{1, i + 2}$$

We also want to ensure periodicity, that is, $ \nabla_{\mathbf{p}_{0, i}}$ and $\nabla_{\mathbf{p}_{0, i}}$ also rely on the points from the previous segment $i - 1$. Then we can write generally for segment $i$ that 

$$\nabla_{\mathbf{p}_{0, i}} \tilde{F} = 12 \mathbf{p}_{0, i - 1} + 192 \mathbf{p}_{0, i} - 72 \mathbf{p}_{1, i - 1} - 144 \mathbf{p}_{1, i} + 12 \mathbf{p}_{0, i + 1}$$

$$\nabla_{\mathbf{p}_{1, i}} \tilde{F} = -144 \mathbf{p}_{0, i} + 36 \mathbf{p}_{1, i - 1} + 144 \mathbf{p}_{1, i} - 72 \mathbf{p}_{0, i + 1} + 36 \mathbf{p}_{1, i + 1}$$

for $i = 1, ..., m.$ For segment $1$, the previous segment that it relies on equals $m$, to ensure periodicity. 

Then we have sketched how to find the gradient $\nabla F = (\nabla_{\mathbf{p}_{0, i}} \tilde{F}, \nabla_{\mathbf{p}_{1, i}} \tilde{F}, ..., \nabla_{\mathbf{p}_{0, m}} \tilde{F}, \nabla_{\mathbf{p}_{1, m}} \tilde{F})$.

Using this information, the gradient of $F$ can be written as

$\nabla F = 
\begin{pmatrix}
12 \mathbf{p}_{0, m} + 192 \mathbf{p}_{0, 1} - 72 \mathbf{p}_{1, m} - 144 \mathbf{p}_{1, 1} + 12 \mathbf{p}_{0, 2} \\
-144 \mathbf{p}_{0, 1} + 36 \mathbf{p}_{1, m} + 144 \mathbf{p}_{1, 1} - 72 \mathbf{p}_{0, 2} + 36 \mathbf{p}_{1, 2} \\
... \\
12 \mathbf{p}_{0, i - 1} + 192 \mathbf{p}_{0, i} - 72 \mathbf{p}_{1, i - 1} - 144 \mathbf{p}_{1, i} + 12 \mathbf{p}_{0, i + 1} \\
-144 \mathbf{p}_{0, i} + 36 \mathbf{p}_{1, i - 1} + 144 \mathbf{p}_{1, i} - 72 \mathbf{p}_{0, i + 1} + 36 \mathbf{p}_{1, i + 1} \\
...\\
12 \mathbf{p}_{0, m - 1} + 192 \mathbf{p}_{0, m} - 72 \mathbf{p}_{1, m - 1} - 144 \mathbf{p}_{1, m} + 12 \mathbf{p}_{0, 1} \\
-144 \mathbf{p}_{0, m} + 36 \mathbf{p}_{1, m - 1} + 144 \mathbf{p}_{1, m} - 72 \mathbf{p}_{0, 1} + 36 \mathbf{p}_{1, 1} 
\end{pmatrix}^T$



### c)

Now, we look at the first order optimality conditions to minimize $\tilde{F}$. If $\mathbf{x}^*$ is a local minimizer, then $\nabla \tilde{F}(\mathbf{x}^*) = 0$.

$$24\mathbf{q}_0 - 36\mathbf{q}_1 + 12\mathbf{q}_3 = 0$$

$$-36\mathbf{q}_0 + 72\mathbf{q}_1 - 36\mathbf{q}_2 = 0$$

$$-36\mathbf{q}_1 + 72\mathbf{q}_2 - 36\mathbf{q}_3 = 0$$

$$12\mathbf{q}_0 - 36\mathbf{q}_2 + 24\mathbf{q}_3 = 0$$

Solving this, we have that $\mathbf{q}_0 = 2\mathbf{q}_2 - 2\mathbf{q}_3$ and $\mathbf{q}_1 = 2\mathbf{q}_2 - \mathbf{q}_3$. Then a minimizer is $\mathbf{x}^* = [2\mathbf{q}_2 - 2\mathbf{q}_3, 2\mathbf{q}_2 - \mathbf{q}_3, \mathbf{q}_2, \mathbf{q}_3]$. This means that we can choose $\mathbf{q}_2$ and $\mathbf{q}_3$ freely, but $\mathbf{q}_0$ and $\mathbf{q}_1$ are then given. Then $\nabla \tilde{F}$ has $2$ degrees of freedom and therefore the solution is not unique. 

Further, we consider the overall problem $F$. Now, we also have to take into account the properties between segments from $C^1$, $\mathbf{p}_{3, i} = \mathbf{p}_{0, i + 1}$ and $\mathbf{p}_{2, i} = 2 \mathbf{p}_{0, i + 1} - \mathbf{p}_{1, i + 1}$. Then each point in segment $i = 1, ..., m - 1$ is given by the points $\mathbf{p}_{0, i + 1}$ and $\mathbf{p}_{1, i + 1}$ in the next segment. So the only points we can choose freely are in the last segment. This segment is not bounded by another segment, since there is no segment $m + 1$. Then we can only choose $\mathbf{p}_{2, m}$ and $\mathbf{p}_{3, m}$ freely, just like we could for $\nabla \tilde{F}$. The remaining control points are then given. Thus $\nabla F$ also has $2$ degrees of freedom, and the solution is not unique either.

### d)

We want to minimize the following expression

$$
\tilde{G}(\mathbf{q}_0, \mathbf{q}_1, \mathbf{q}_2, \mathbf{q}_3) = \frac{\lambda}{2}(||\mathbf{s}- \mathbf{q}_0||_{2}^{2} + ||\mathbf{e}- \mathbf{q}_3||_{2}^{2}) + \tilde{F}(\mathbf{q}_0, \mathbf{q}_1, \mathbf{q}_2, \mathbf{q}_3)
$$

Like we did in 4c), lets find such a minimizer so that $\nabla \tilde{G}_\lambda(\mathbf{x}^*) = 0$.

Lets find an expression for the derivative of $\tilde{G}_\lambda$.

$$
\nabla \tilde{G}_\lambda(\mathbf{q}_0, \mathbf{q}_1, \mathbf{q}_2, \mathbf{q}_3) = (\frac{\partial}{\partial \mathbf{q}_0} \tilde{G}_\lambda, \frac{\partial}{\partial \mathbf{q}_1} \tilde{G}_\lambda, \frac{\partial}{\partial \mathbf{q}_2} \tilde{G}_\lambda, \frac{\partial}{\partial \mathbf{q}_3} \tilde{G}_\lambda)
$$

We already know what $\nabla \tilde{F}$ is from 4b), so lets calculate the partial derivatives for the first term

$$
\frac{\partial}{\partial \mathbf{q}_0}(\frac{\lambda}{2}(||\mathbf{s}- \mathbf{q}_0||_{2}^{2} + ||\mathbf{e}- \mathbf{q}_3||_{2}^{2}))
\\
\;
\\
\frac{\partial}{\partial \mathbf{q}_1}(\frac{\lambda}{2}(||\mathbf{s}- \mathbf{q}_0||_{2}^{2} + ||\mathbf{e}- \mathbf{q}_3||_{2}^{2})) = 0
\\
\;
\\
\frac{\partial}{\partial \mathbf{q}_2}(\frac{\lambda}{2}(||\mathbf{s}- \mathbf{q}_0||_{2}^{2} + ||\mathbf{e}- \mathbf{q}_3||_{2}^{2})) = 0
\\
\;
\\
\frac{\partial}{\partial \mathbf{q}_3}(\frac{\lambda}{2}(||\mathbf{s}- \mathbf{q}_0||_{2}^{2} + ||\mathbf{e}- \mathbf{q}_3||_{2}^{2}))
$$

Lets calculate the first and the last derivative. For simplicity, lets define $A = \mathbf{s}- \mathbf{q}_0$ and $B = \mathbf{e}- \mathbf{q}_3$. 

$$
|| A ||_{2}^{2} = (\sqrt{A^2})^2 = A^2 = (\mathbf{s}- \mathbf{q}_0)(\mathbf{s}- \mathbf{q}_0) = \mathbf{s}^2 -2\mathbf{s}\mathbf{q}_0 + \mathbf{q}_0^2
\\
\;

|| B ||_{2}^{2} = (\sqrt{B^2})^2 = B^2 = (\mathbf{e}- \mathbf{q}_3)(\mathbf{e}- \mathbf{q}_3) = \mathbf{e}^2 -2\mathbf{e}\mathbf{q}_3 + \mathbf{q}_3^2
$$
$$
\frac{\partial}{\partial \mathbf{q}_0}(\mathbf{s}^2 -2\mathbf{s}\mathbf{q}_0 + \mathbf{q}_0^2 + \mathbf{e}^2 -2\mathbf{e}\mathbf{q}_3 + \mathbf{q}_3^2) = -2\mathbf{s} + 2\mathbf{q}_0
\\
\;
\\
\frac{\partial}{\partial \mathbf{q}_3}(\mathbf{s}^2 -2\mathbf{s}\mathbf{q}_0 + \mathbf{q}_0^2 + \mathbf{e}^2 -2\mathbf{e}\mathbf{q}_3 + \mathbf{q}_3^2) = -2\mathbf{e} + 2\mathbf{q}_3
$$

Putting everything together we get the following derivative

$$
\nabla \tilde{G}_\lambda(\mathbf{q}_0, \mathbf{q}_1, \mathbf{q}_2, \mathbf{q}_3) = 
\begin{pmatrix}

-36\mathbf{q}_0 +72\mathbf{q}_1 -36\mathbf{q}_2 \\

\end{pmatrix}
$$

A minimizer for $\nabla \tilde{G}_\lambda$ is given by

$$-\lambda \mathbf{s} + (\lambda + 24)\mathbf{q}_0 - 36\mathbf{q}_1 + 12\mathbf{q}_3 = 0$$
$$-36\mathbf{q}_0 +72\mathbf{q}_1 -36\mathbf{q}_2 = 0$$
$$-36\mathbf{q}_1 +72\mathbf{q}_2 -36\mathbf{q}_3 = 0$$
$$-\lambda \mathbf{e} + 12\mathbf{q}_0 - 36\mathbf{q}_2 + (\lambda + 24)\mathbf{q}_3 = 0$$

Choosing $\lambda = 1$. Solving the equations, we have that 

$$\mathbf{s} = \mathbf{q}_0$$ 
$$\mathbf{e} = \mathbf{q}_3$$
$$\mathbf{q}_1 = \frac{1}{3}\mathbf{e} + \frac{2}{3}\mathbf{s}$$ 
$$\mathbf{q}_2 = \frac{2}{3}\mathbf{e} + \frac{1}{3}\mathbf{s}$$

Then a minimizer for $\nabla \tilde{G}_\lambda$ is $\mathbf{x}^* = [\mathbf{s}, \frac{1}{3}\mathbf{e} + \frac{2}{3}\mathbf{s}, \frac{2}{3}\mathbf{e} + \frac{1}{3}\mathbf{s}, \mathbf{e}]$. Which means that each point is given based on $\mathbf{s}$ and $\mathbf{e}$. Then we cant choose any point freely, and the solution is unique. 

$\lambda$ determines the weighting between our data points. It changes how smooth the graph is depending on the value. If we let $\lambda \rightarrow 0$, the graph becomes smoother and smoother because the weighting of the data points becomes less important, so that the distance between the curve and the data points increases. If we however let $\lambda \rightarrow \infty$, the weighting between the data points increases and become more important, so that the curve should not loose its distance to these points. What happens to the distance from the curve to the control points, is something we will explore further in the next exercise. 

We want to find the gradient of $G_\lambda$. We have that

We have the gradient of $F$ and we have the gradient of the first summand as well from $\tilde{G}$. Lets write it generally

$$\nabla_{\mathbf{p}_{1, i}} \tilde{G}_\lambda = 0$$

We can extend this into an expression for $\nabla G_\lambda$

Thus $G_\lambda$ is also unique. $\lambda$ doesn't change the uniqueness as long as it is $> 0$. It simply determines the shape of the curve we get.
```python
def gGrad(lam,p,d):
    '''
    Input: Lambda parameter, control points p, data points d
    Output: Gradient G
    '''
    
    m = p.shape[0]            # Number of splines
    
    grad =  np.zeros((m,2,2)) # Zero-array to store the gradient in
    
    # Exception case for the first spline because of periodicity
    grad[0][0] = 12*p[-1][0]+192*p[0][0]-72*p[-1][1]-144*p[0][1]+12*p[1][0] + lam*(p[0][0]-d[0])
    grad[0][1] = - 144*p[0][0]+36*p[-1][1]+144*p[0][1]-72*p[1][0]+36*p[1][1]

      
    for i in range(1,m-1): # Iterate through all but the first and last splines
        grad[i][0] = 12*p[i-1][0]+192*p[i][0]-72*p[i-1][1]-144*p[i][1]+12*p[i+1][0] +lam*(p[i][0]-d[i])     
        grad[i][1] = - 144*p[i][0]+36*p[i-1][1]+144*p[i][1]-72*p[i+1][0]+36*p[i+1][1]
 
    
    # Exception case for last spline, also considering periodicity
    grad[-1][0] = 12*p[-2][0]+192*p[-1][0]-72*p[-2][1]-144*p[-1][1]+12*p[0][0] + lam*(p[-1][0]-d[-1])
    grad[-1][1] = - 144*p[-1][0]+36*p[-2][1]+144*p[-1][1]-72*p[0][0]+36*p[0][1]
    
    return grad
```

```python
def steepestDescent(grad, x0, a, lam, data, tol, maxIter):
    '''
    Action: Uses the steepest descent method to minimize the distance between the curve and control points

    Input:
    grad    - Gradient function
    x0      - Initial control points
    a       - Step size
    lam     - Lambda parameter
    data    - Data points to consider
    tol     - Tolerance for when to stop
    maxIter - Maximum number of iterations

    Output: Matrix with all versions of the control points
    '''
    m = x0.shape[0]                    # Number of splines
    xVals = np.zeros((maxIter,m,2,2))  # Zero matrix to store all versions in
    xVals[0] = x0                      # Store original versions of points first
    x = x0

    for k in range(1, maxIter):        # Iterate until maxIter is reached or tolerance is ok
        g  = grad(lam,xVals[k-1],data) # Compute gradient of the last version of points wrt. data points
        d = -g
        xNew = x + a*d                 # Update points using gradient descent method
        xVals[k] = xNew                # Store these new points
        x = xNew

        # Check if the tolerance is reached
        if np.linalg.norm(xVals[k] - xVals[k-1]) < tol:
            print(f'Tolerance between x values reached after {k} iterations')
            break

    xVals = xVals[:k]                  # Remove unused space in matrix
    if k == maxIter-1:
        print("Maximum number of iterations reached")

    return xVals
```

```python
def restore(p,v):
    '''
    Action: Restores points based on previously calculated coherence between points
    Input:  Stripped points, velocities of original points
    Output: Restored points
    '''
    m = p.shape[0]           # Number of splines
    pNew = np.zeros((m,4,2)) # Zero array to store points in

    for i in range(m-1):     # Iterating through the splines
        # Recovering points
        p0i = p[i][0]
        p1i = p[i][1]
        p2i = p[i+1][0]-v[i+1]/3
        p3i = p[i+1][0]
        
        pNew[i] = np.array([p0i,p1i,p2i,p3i])
    
    # Edge case for last spline to ensure periodicity
    p0m = p[m-1][0]
    p1m = p[m-1][1]
    p2m = p[0][0]-v[0]/3
    p3m = p[0][0]
        
    pNew[m-1] = np.array([p0m,p1m,p2m,p3m])
    return pNew
```

```python
def strip(p):
    '''
    Action: Removes unwanted control points
    Input:  All points
    Output: Stripped points
    '''
    m = p.shape[0]            # Number of splines
    index = [4,5,6,7]         # Indexes where unwanted points are
    pNew = np.zeros((m,2,2))  # Zero array to store points in

    for i in range(m):        # Iterating through splines
        temp = p[i].flatten() # Flatten points to easily remove unwanted points
        news = np.delete(temp,index).reshape(2,2) # Remove and reshape points
        pNew[i] = news
        
    return pNew
```

With a value of $\lambda=5$, it makes sense that we never reach our data points because they are of less importance. This is something that we can clearly observe in the last plot after many iterations for $\lambda=5$. We observe that when $\lambda$ increases, the graph looses less distance to the data points after many iterations, which is because of the importance of the data points, as we discussed before. Consequently, it becomes less smooth as we can see in the last plot.
